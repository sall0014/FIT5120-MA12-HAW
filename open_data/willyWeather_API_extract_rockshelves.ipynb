```python
#libraries
import requests
import json
import pandas as pd
import datetime
```

```python
rockshelves_id = [('Cape Woolamai', '11840'), ('Kilcunda', '12454'), 
                  ('Lorne', '12614'), ('Portland', '13034'), 
                  ('Point Lonsdale', '13014'), ('Cape Schanck', '13813'), 
                  ('Flinders', '13832'), ('Tongue Point', '35555'), ('Point Roadknight','27126')]
```

```python
def get_api(id):
    url = "https://api.willyweather.com.au/v2/N2FkOTJkOWFkZDA4NWIzOTZlODhkOD/locations/" + id + "/weather.json?forecasts=tides&days=200&startDate=2021-04-18"
    resp = requests.get(url)
    rockshelf = json.loads(resp.text)
    return rockshelf
```

```python
date = []
time = []
height = []
tide_type = []
location_id = []
location_name = []

for rockshelf_id in rockshelves_id:
    rockshelf_details = get_api(rockshelf_id[1])
    #get each date
    dates = rockshelf_details['forecasts']['tides']['days']
  
    for dat in dates:
        for day in dat['entries']:
            date_time = datetime.datetime.strptime(day['dateTime'], "%Y-%m-%d  %H:%M:%S")
            date.append(date_time.date())
            time.append(date_time.time())
            height.append(day['height'])
            tide_type.append(day['type'])
            location_id.append(rockshelf_id[1])
            location_name.append(rockshelf_id[0])
```

```python
#create datframe
date_tide = pd.DataFrame(columns = ['date', 'time', 'height', 'type', 'location_id', 'location_name'])
#add data to dataframe
date_tide['date'] = date
date_tide['time'] = time
date_tide['height'] = height
date_tide['type'] = tide_type
date_tide['location_id'] = location_id
date_tide['location_name'] = location_name
```

```python
date_tide.insert(0, 'ID', date_tide.index)
date_tide['ID'] = 'Tide' + date_tide['ID'].astype('str')
```
